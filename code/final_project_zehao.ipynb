```python
def report_tesult(y_test,predicted_target):
    print('precision',sklearn.metrics.precision_score(y_test, predicted_target))
    print('f1',f1_score(y_test, predicted_target, average='weighted'))
    print('recall',recall_score(y_test, predicted_target, average='weighted'))
    print('accur',sklearn.metrics.balanced_accuracy_score(y_test, predicted_target))
    print('auc',roc_auc_score(y_test, predicted_target))
    tn, fp, fn, tp = confusion_matrix(y_test, predicted_target).ravel()
    specificity = tn / (tn+fp)
    print('spe',specificity)
```

```python
def get_aucmap_and_spe_1(clf,X_train_set,X_test_set,y_train_set,y_test_set):
    clf.fit(X_train_set, y_train_set)
    pred_result = clf.predict_proba(X_test_set)
    idx = list(filter(lambda x: x==1, clf.classes_))[0]
    #print(idx)
    pred_result_1 = list(map(lambda x: x[idx], pred_result))
    #print(pred_result_1)
    fpr, tpr, thresholds = roc_curve(y_test_set, pred_result_1)
    #print(pred_result_1)
    optidx = youden_J_score(fpr, tpr)
    pred_result_1 = np.array(pred_result_1, dtype=np.float32)
    sen,spe=get_sensitivity_specificity(tpr, fpr, optidx)
    roc_auc = auc(fpr, tpr)
    lw = 2
    plt.plot(fpr, tpr, color='darkorange',
             lw=lw, label='ROC curve (area = {:0.4f})'.format(auc(fpr, tpr)))
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve -- GB (without feature selection)')
    plt.legend(loc="lower right")
    plt.show()
    print(roc_auc_score(y_test_set, pred_result_1))
    print('sensitivity:',sen)
    print('specificity',spe)
```

```python


def five_fold_CV(clf, params, x, y):
    cv_model = RandomizedSearchCV(clf, params, scoring='roc_auc', n_jobs=-1,
                                  cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=2020), 
                                  verbose=1,n_iter=30, refit=True)
    
    cv_model.fit(x, y)
    
    return cv_model.best_estimator_, cv_model.best_score_
```
